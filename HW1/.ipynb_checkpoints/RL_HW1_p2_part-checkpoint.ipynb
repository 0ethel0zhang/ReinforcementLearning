```python
import numpy as np
from functools import reduce
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def cap200(x):
    """cap number of passengers waiting to 200"""
    new = []
    for num in x:
        if num > 200:
            new.append(200)
        else:
            new.append(num)
    return new
```

```python
def count(x):
    "count occurrences in a list"
    dic = {}
    if type(x) == int:
        dic[x] = 1
    else:
        for t in x:
            if t in dic.keys():
                dic[t] += 1
            else:
                dic[t] = 1
    return dic
```

## 1. a) Enumeration

```python
def enumeration(T):
    """enumeration function"""
    if T==0:
        return [[wp], 1]
    else:
        #previous state
        last_state = enumeration(T-1)
        last_p = last_state[1]
        last_wp = last_state[0]
        #current state
        left = [p + x for x in range(1,6) for p in last_wp]
        wp_counter = cap200(left)
        result_counter = count(wp_counter)
        wp_counter_2 = [t[0]for t in result_counter.items()]
        total = reduce((lambda x, y: x + y), [t[1]for t in result_counter.items()])
        p = [t[1] / total for t in result_counter.items()]
        #update state
        out_0 = sum(list(map(lambda x: x[0]*x[1], list(zip([-2 * x for x in wp_counter_2], p)))))
        out_1 = sum(list(map(lambda x: x[0]*x[1], list(zip([-2 * x for x in [m - min(m, 15) for m in wp_counter_2]], p))))) - 100
        #out_0 = sum([-2 * p * x for x in wp_counter])
        #out_1 = sum([-2 * (t-min(15,t)) * p for t in wp_counter]) - 100
        if max(out_0,out_1) == out_1:
            #print("optimal action at T{} is: {}; # of waiting passengers: ; reward is: {}".format(T, 0, out_1))
            value[T] = out_1
            return [[x - min(x, 15) for x in wp_counter_2], p]
        else:
            #left = (left - min(15, left))
            #print("optimal action at T{} is: {}; # of waiting passengers: ; reward is: {}".format(T, 1, out_0))
            value[T] = out_0
            return [wp_counter_2, p]
```

```python
values = []
for wp in range(0,201):
    a, b = enumeration(500)
    V0 = 0
    for k, v in value.items():
        V0 += v * (0.95 ** k)
    values.append(V0)
```

```python
plt.plot(values)
plt.xlabel("Number of Passengers Waiting at T0")
plt.ylabel("Optimal Value")
plt.title("Number of Passengers Waiting at T0 vs. Optimal Value")
```

## 1. b) Value Iteration

```python
def valueIteration(wp, V_init = 0, p = 1):
    left = [wp + x for x in range(1,6)]
    wp_counter = cap200(left)
    #result_counter = count(left)
    #wp_counter = [t[0]for t in result_counter.items()]
    #total = reduce((lambda x, y: x + y), [t[1]for t in result_counter.items()])
    #p = [t[1] / total for t in result_counter.items()]
    p = 1/5
    #update state
    out_0 = sum([(-2 * x + 0.95 * V_init) * p for x in wp_counter])
    out_1 = sum([(-2 * x + 0.95 * V_init)* p for x in [m - min(m, 15) for m in wp_counter]]) - 100
    #print(out_0, out_1)
    out = max(out_0, out_1)
    delta = abs(out - V_init)
    #print(delta)
    if delta > 0.05:
        valueIteration(wp, out, p)
    else:
        value.append(out)
        return out
```

```python
value = []
for wp in range(0, 201):
    valueIteration(wp)
```

```python
plt.plot(value)
plt.xlabel("Number of Passengers Waiting at T0")
plt.ylabel("Optimal Value")
plt.title("Number of Passengers Waiting at T0 vs. Optimal Value")
```
