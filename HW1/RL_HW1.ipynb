## 1. a) Enumeration

```python
def enumeration(T):
    """enumerate T times for optimal value"""
    Vt = np.zeros(201)
    Vt1 = np.zeros(201)
    for t in range(T, -1, -1):
        for wp in range(201):
            V_out0 = -2 * wp
            V_out1 = -100 + (-2) * (wp - min(wp, 15))
            for new_p in range(1, 6):
                V_out0 +=  1/5 * .95 * Vt1[min(wp + new_p, 200)]
                V_out1 += 1/5 * .95 * Vt1[(wp + new_p)- min((wp + new_p), 15)]
            Vt[wp] = max(V_out0, V_out1)
        Vt1 = Vt
    return Vt
```

```python
Vt = enumeration(500)
```

## 1. b) Value Iteration

```python
def valueIteration(guess = 0, threshold = 0.01):
    """Start iteration with an initial guess"""
    V_init = np.array([float(guess)] * 201)
    while True: 
        V_new = V_init.copy()
        delta = 0
        for wp in range(201):
            V_out0 = -2 * wp
            V_out1 = -100 + (-2) * (wp - min(wp, 15))
            for new_p in range(1, 6):
                V_out0 += 1/5 * 0.95 * V_new[min(wp + new_p, 200)]
                V_out1 += 1/5 * 0.95 * V_new[(wp + new_p)- min((wp + new_p), 15)]
            V_init[wp] = max(V_out0, V_out1)
        delta = max(delta, max(abs(V_new-V_init)))
        if delta < threshold:
            return V_init
            break
```

## 1. c) Policy Iteration

```python
def policyIteration(V_guess = 0, p_guess = 1, threshold = 0.01):
    """policy updates with initial guesses"""
    # all possible actions
    a = [0,1]
    # arbitrary values
    V_init = np.array([V_guess] * 201)
    V_new = V_init.copy()
    # arbitrary policies
    policy = np.array([p_guess] * 201)
    a.remove(p_guess)
    policy_new = np.array([a[0]] * 201)
    # initialize delta as a large #
    delta_init = 100
    delta = delta_init
    
    changed = True
    while changed: 
        # value iteration
        while threshold < delta:
            V_new = V_init.copy()
            for wp in range(201):
                delta = 0
                v = V_init[wp] # old value
                V_out0 = -2 * wp
                V_out1 = -100 + (-2) * (wp - min(wp, 15))
                for new_p in range(1, 6):
                    V_out0 += 1/5 * 0.95 * V_new[min(wp + new_p, 200)]
                    V_out1 += 1/5 * 0.95 * V_new[(wp + new_p)- min((wp + new_p), 15)]
                values = [V_out0, V_out1]
                V_init[wp] = values[policy[wp]]
                delta = max(delta, abs(v-V_init[wp]))

        # policy improvement
        for wp in range(201):
            V_out1 = -100 + (-2) * (wp - min(wp, 15))
            V_out0 = -2 * wp
            for new_p in range(1, 6):
                V_out1 += 1/5 * 0.95 * V_new[(wp + new_p)- min((wp + new_p), 15)]
                V_out0 += 1/5 * 0.95 * V_new[min(wp + new_p, 200)]
            values = [V_out0, V_out1]
            policy_new[wp] = np.argmax(values)

        # change policy if not the same
        if sum(policy - policy_new) == 0:
            changed = False
        else:
            delta = delta_init
            policy = policy_new.copy()
            
    return policy
```
